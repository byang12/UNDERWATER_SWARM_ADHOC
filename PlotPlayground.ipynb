```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Create some sample data (e.g., a correlation matrix)
data = np.array([
    [9569.1,5553.5,3800.2,2613.7,1693.7,1089.7],
    [1253.2,522.3,473.1,456.8,426.7,399.1],
    [516.3,502.2,411.5,475.5,466.7,464.8],
    [515.5,469.8,441.4,437.8,410.1,407.5],
    [503.8,473.8,441.7,416.9,411.6,403.5],
    [474.1,461.4,441.7,411,409.2,400.9]
])
df = pd.DataFrame(data, columns=[f'{(i+5)**2}' for i in range(6)],
                  index=[f'{100+i*100} m' for i in range(6)])
```

```python
# Create the heatmap
plt.figure(figsize=(8, 6)) # Adjust figure size for better readability
sns.heatmap(df, annot=True, vmin=400, vmax=1500,cmap='viridis', fmt=".2f", linewidths=.5)
plt.xlabel("Number of AUVs", fontsize=12)
plt.ylabel("Transmission Range", fontsize=12)
#plt.title('Sample Heatmap')
plt.show()
```
